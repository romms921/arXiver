```python
import pandas as pd
import os

# Define input files to check
input_files = ["latex_affiliations_output.txt", "latex_affiliations_output_2.txt"]
found_file = None

for f_path in input_files:
    if os.path.exists(f_path):
        found_file = f_path
        break

if not found_file:
    print("Error: No LaTeX affiliation output files found.")
else:
    print(f"Parsing {found_file}...")
    error_titles = []
    current_title = None

    with open(found_file, 'r') as f:
        for line in f:
            if line.startswith("PAPER: "):
                current_title = line.replace("PAPER: ", "").strip()
            elif line.startswith("ERROR: Failed to download LaTeX sources"):
                if current_title:
                    error_titles.append(current_title)
                    current_title = None # Reset to avoid double counting if error repeats (unlikely)

    # Create DataFrame
    df_errors = pd.DataFrame(error_titles, columns=['title'])
    
    print(f"Extracted {len(df_errors)} paper titles with download errors.")
    
    # Display the result
    if not df_errors.empty:
        display(df_errors.head(20))
    else:
        print("No papers with download errors found.")

# Optional: preview df_errors
df_errors
```

```python
# Delete Duplicates
df_errors = df_errors.drop_duplicates()
```

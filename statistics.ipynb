```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.dpi': 250})
plt.rcParams['font.sans-serif'] = 'Times New Roman'
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('FINAL_ARXIV_2025_copy_updated.csv')
```

```python
data
```

```python
# Month Column
data['date'] = pd.to_datetime(data['date'])
data['month'] = data['date'].dt.month
```

```python
# Average Number of Papers per month
avg_month_paper = data.groupby('month').size().mean()
print("Average Number of Papers Per Month:", avg_month_paper.round(0))

# Calculate the total number of papers
total_papers = len(data)

# Calculate the number of days in the range
day_range = (data['date'].max() - data['date'].min()).days + 1

# Calculate average
avg_day_paper = total_papers / day_range
print("Average Number of Papers Per Day:", round(avg_day_paper, 0))
```

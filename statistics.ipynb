```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.dpi': 250})
plt.rcParams['font.sans-serif'] = 'Times New Roman'
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('FINAL_ARXIV_2025_with_affiliations.csv')
```

```python
data
```

```python
# Month Column
data['date'] = pd.to_datetime(data['date'])
data['month'] = data['date'].dt.month
```

```python
# Average Number of Papers per month
avg_month_paper = data.groupby('month').size().mean()
print("Average Number of Papers Per Month:", avg_month_paper.round(0))

# Calculate the total number of papers
total_papers = len(data)

# Calculate the number of days in the range
day_range = (data['date'].max() - data['date'].min()).days + 1

# Calculate average
avg_day_paper = total_papers / day_range
print("Average Number of Papers Per Day:", round(avg_day_paper, 0))
```

```python
# Ranking Months where most papers were published
month_counts = data['month'].value_counts().sort_index()
month_names = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
month_counts.index = month_names
month_ranking = month_counts.sort_values(ascending=False)
print("\nMonths Ranked by Number of Papers Published:")
montly_counts = [count for month, count in month_ranking.items()]
for month, count in month_ranking.items():
    print(f"{month}: {count} papers")



# Accounting for number of days in each month by plotting average papers per day in each month
# Obtain number of days from date column
days_in_month = data.groupby('month')['date'].first().dt.days_in_month
days_in_month.index = month_names
avg_papers_per_day = month_counts / days_in_month
print("\nAverage Papers Published Per Day in Each Month:")
for month, avg in avg_papers_per_day.items():
    print(f"{month}: {avg:.0f} papers/day")
```

```python
std_month_counts = month_counts.std()
print("\nStandard Deviation of Monthly Paper Counts:", round(std_month_counts, 2))

# Number of Sigma between October and March
october_count = month_counts['October']
march_count = month_counts['March']
difference = abs(october_count - march_count)
num_sigma = difference / std_month_counts
print(f"\nNumber of Sigma between October and March: {num_sigma:.2f}")
```

```python
# Average number of papers under 'Astrophysics of Galaxies' category per month
# Group by primary subject and count the number of papers
subject_counts = data['primary_subject'].value_counts()

# Sort in descending order
subject_counts_sorted = subject_counts.sort_values(ascending=False)

print("Subjects Arranged in Descending Order of Number of Papers:")
print(subject_counts_sorted)
```

```python
# Average number of papers under 'Astrophysics of Galaxies' category per month
# Group by primary subject and month, then count
subject_month_counts = data.groupby(['primary_subject', 'month']).size()

# Calculate the average per month for each subject
avg_subject_per_month = subject_month_counts.groupby('primary_subject').mean()

# Sort in descending order
subject_counts_sorted = avg_subject_per_month.sort_values(ascending=False)

print("Subjects Arranged in Descending Order of Average Number of Papers Per Month:")
print(subject_counts_sorted.round(0))
```

```python
# Average number of papers under 'Astrophysics of Galaxies' category per day
# Group by primary subject and count the total number of papers
subject_total_counts = data['primary_subject'].value_counts()

# Calculate the average per day for each subject
# We use the previously calculated 'day_range' variable
avg_subject_per_day = subject_total_counts / day_range

# Sort in descending order
subject_counts_sorted_day = avg_subject_per_day.sort_values(ascending=False)

print("Subjects Arranged in Descending Order of Average Number of Papers Per Day:")
print(subject_counts_sorted_day.round(0))
```

```python
# Which month had the highest number of papers for each Primary Subject
highest_month_per_subject = data.groupby(['primary_subject', 'month']).size().reset_index(name='count')
highest_month_per_subject = highest_month_per_subject.loc[highest_month_per_subject.groupby('primary_subject')['count'].idxmax()]
highest_month_per_subject['month_name'] = highest_month_per_subject['month'].apply(lambda x: month_names[x-1])

# Sort in descending order of count
highest_month_per_subject = highest_month_per_subject.sort_values(by='count', ascending=False)

print("Month with Highest Number of Papers for Each Primary Subject:")
print(highest_month_per_subject[['primary_subject', 'month_name', 'count']])
```

```python
# Average Citation Count for each Primary Subject
avg_citation_per_subject = data.groupby('primary_subject')['total_citations'].mean().sort_values(ascending=False)
print("Average Citation Count for Each Primary Subject:")
print(avg_citation_per_subject.round(2))

# Average Citation Count for each Primary Subject removing self-citations
avg_citation_no_self_per_subject = data.groupby('primary_subject')['non_self_citations'].mean().sort_values(ascending=False)
print("Average Citation Count (No Self-Citations) for Each Primary Subject:")
print(avg_citation_no_self_per_subject.round(2))
```

```python
# Only considering the papers with a journal_flag
data_journal = data[data['journal_flag'] == 1]
# Average Citation Count for each Primary Subject
avg_citation_per_subject = data_journal.groupby('primary_subject')['total_citations'].mean().sort_values(ascending=False)
print("Average Citation Count for Each Primary Subject:")
print(avg_citation_per_subject.round(2))

# Average Citation Count for each Primary Subject removing self-citations
avg_citation_no_self_per_subject = data_journal.groupby('primary_subject')['non_self_citations'].mean().sort_values(ascending=False)
print("Average Citation Count (No Self-Citations) for Each Primary Subject:")
print(avg_citation_no_self_per_subject.round(2))
```

```python
import ast

# Secondary Subject Statistics
# Print Unique Secondary Subjects
# Clean and split the secondary subjects
# The column contains string representations of lists, so we need to parse them

# Function to parse the stringified lists
def parse_subjects(subject_str):
    try:
        return ast.literal_eval(subject_str)
    except (ValueError, SyntaxError):
        return []

# Apply parsing and explode the list to get individual subjects
all_secondary_subjects = data['secondary_subjects'].dropna().apply(parse_subjects).explode()
unique_secondary_subjects = all_secondary_subjects.unique()

print("Unique Secondary Subjects (Individual):")
for subject in unique_secondary_subjects:
    print(subject)


# Count occurrences of each secondary subject in descending order
secondary_subject_counts = all_secondary_subjects.value_counts()
print("\nSecondary Subjects Ranked by Number of Papers:")
for subject, count in secondary_subject_counts.items():
    print(f"{subject}: {count} papers")
```

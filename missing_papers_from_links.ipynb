```python
# Imports
import pandas as pd
import numpy as np
import urllib.request as libreq
from groq import Groq
import xml.etree.ElementTree as ET
import certifi
import os
from datetime import datetime
from tqdm import tqdm
from dotenv import load_dotenv
load_dotenv()
from bs4 import BeautifulSoup
import re
import PyPDF2
import io
import warnings
warnings.filterwarnings('ignore')

os.environ['SSL_CERT_FILE'] = certifi.where()

# Load the previous data
prev_df = pd.read_csv('datasets/arxiv_papers.csv')
prev_non_existent = pd.read_csv('datasets/non_existent.csv')
non_existent_dates = []
non_existent_titles = []

# Define default date for error handling
paper_date = datetime.now().strftime('%Y-%m-%d')

# Define function to extract metadata from single abstract page
def extract_single_paper_metadata(soup, arxiv_id):
    # Title
    title_tag = soup.find('h1', class_='title mathjax')
    title = title_tag.get_text(strip=True).replace('Title:', '').strip() if title_tag else None

    # Abstract
    abstract_tag = soup.find('blockquote', class_='abstract mathjax')
    abstract = abstract_tag.get_text(strip=True).replace('Abstract:', '').strip() if abstract_tag else None

    # Authors
    authors_tag = soup.find('div', class_='authors')
    authors = [author.get_text(strip=True) for author in authors_tag.find_all('a')] if authors_tag else []

    # Comments and Journal ref
    comments = ''
    published_journal = None
    
    comments_tag = soup.find('td', class_='tablecell comments mathjax')
    if comments_tag:
        comments = comments_tag.get_text(strip=True)
        
    journal_tag = soup.find('td', class_='tablecell jref')
    if journal_tag:
        published_journal = journal_tag.get_text(strip=True)

    # Figures, Pages, Tables
    figures_match = re.search(r'(\d+)\s+figures', comments)
    figures = int(figures_match.group(1)) if figures_match else None
    pages_match = re.search(r'(\d+)\s+pages', comments)
    pages = int(pages_match.group(1)) if pages_match else None
    tables_match = re.search(r'(\d+)\s+table[s]?', comments)
    tables = int(tables_match.group(1)) if tables_match else None

    # PDF link
    pdf_link = f'arxiv.org/pdf/{arxiv_id}'

    # Subjects
    primary_subject = None
    secondary_subjects = None
    
    subjects_tag = soup.find('td', class_='tablecell subjects')
    if subjects_tag:
        primary_subject_tag = subjects_tag.find('span', class_='primary-subject')
        if primary_subject_tag:
            primary_subject = primary_subject_tag.get_text(strip=True)
            # Remove primary subject from text to find secondary
            full_text = subjects_tag.get_text(strip=True)
            subjects_split = full_text.split(';')
            secondary_temp = [s.strip() for s in subjects_split if primary_subject not in s]
            if secondary_temp:
                secondary_subjects = secondary_temp
        else:
            primary_subject = subjects_tag.get_text(strip=True)

    # Journal parsing logic
    submitted_journal = comments.split('Submitted to ')[-1] if 'Submitted to' in comments else None
    submitted_journal = comments.split('Accepted to ')[-1] if 'Accepted to' in comments else submitted_journal
    submitted_journal = comments.split('Accepted for publication in ')[-1] if 'Accepted for publication in' in comments else submitted_journal
    submitted_journal = comments.split('Accepted by ')[-1] if 'Accepted by' in comments else submitted_journal
    submitted_journal = comments.split('Submitted by ')[-1] if 'Submitted by' in comments else submitted_journal

    # Submisison Date
    current_date = paper_date
    dateline = soup.find('div', class_='dateline')
    if dateline:
        match = re.search(r'Submitted on (\d{1,2} \w+ \d{4})', dateline.get_text())
        if match:
            try:
                dt = datetime.strptime(match.group(1), '%d %b %Y')
                current_date = dt.strftime('%Y-%m-%d')
            except:
                pass

    return {
        'title': title,
        'abstract': abstract,
        'authors': authors,
        'figures': figures,
        'pages': pages,
        'tables': tables,
        'pdf_link': pdf_link,
        'primary_subject': primary_subject,
        'secondary_subjects': secondary_subjects,
        'submitted_journal': submitted_journal,
        'published_journal': published_journal,
        'date': current_date
    }

# Retrieve Metadata for missing papers
metadata_list = []
print(f"Retrieving metadata for {len(missing_papers)} papers...")

for i, row in tqdm(missing_papers.iterrows(), total=len(missing_papers), desc="Fetching metadata"):
    arxiv_id = row['arxiv_id']
    url = f'https://arxiv.org/abs/{arxiv_id}'
    
    try:
        page = libreq.urlopen(url)
        html = page.read().decode('utf-8')
        soup = BeautifulSoup(html, 'html.parser')
        
        metadata = extract_single_paper_metadata(soup, arxiv_id)
        metadata_list.append(metadata)
        
        # update global paper_date to the last successfully fetched one (for compatibility)
        paper_date = metadata['date'] 
        
    except Exception as e:
        print(f"Failed to retrieve {arxiv_id}: {e}")
        non_existent_dates.append(paper_date)
        non_existent_titles.append(f"ID: {arxiv_id}")

# Dataframe conversion function
def metadata_to_dataframe(metadata_list):
    return pd.DataFrame(metadata_list)

# Create DataFrame
df = metadata_to_dataframe(metadata_list)
print('Retrieved all Metadata')

# Remove brackets
def remove_brackets(text):
    if text is None: return None
    return re.sub(r'\(.*?\)', '', text).strip()

# Data preprocessing
if not df.empty:
    df['primary_subject'] = df['primary_subject'].map(remove_brackets)
    df['secondary_subjects'] = df['secondary_subjects'].map(lambda x: [remove_brackets(subject) for subject in x] if x is not None else None, na_action='ignore') 
    df['submitted_journal'] = df['submitted_journal'].str.split(r'[,;:.]').str[0]

# For loop to retrieve missing metadata (Pages/Figures/Tables from PDF)
for i in tqdm(range(len(df)), desc="Retrieving missing metadata"):
    try:
        if pd.isna(df['pages'][i]) or pd.isna(df['figures'][i]) or pd.isna(df['tables'][i]):

            pdf_link = df['pdf_link'][i]
            # Ensure proper URL format
            if not pdf_link.startswith('http'):
                url_to_fetch = 'https://' + pdf_link
            else:
                url_to_fetch = pdf_link
                
            pdf_response = libreq.urlopen(url_to_fetch)
            pdf_file = pdf_response.read()
            # print(f"Processing PDF: {pdf_link}")
            pdf_reader = PyPDF2.PdfReader(io.BytesIO(pdf_file))

            # Number of Pages
            if pd.isna(df['pages'][i]):
                num_pages = len(pdf_reader.pages)
                df.at[i, 'pages'] = num_pages

            # Number of Figures
            if pd.isna(df['figures'][i]):
                highest_figure_number = 0
                for page in pdf_reader.pages:
                    text = page.extract_text()
                    figure_numbers = re.findall(r'(?i)(?:Figure|Fig.|Figure.|Fig})\s+(\d+)', text)
                    if figure_numbers:
                        highest_figure_number = max(highest_figure_number, max(map(int, figure_numbers)))
                df.at[i, 'figures'] = highest_figure_number

            # Number of Tables
            if pd.isna(df['tables'][i]):
                highest_table_number = 0
                for page in pdf_reader.pages:
                    text = page.extract_text()
                    table_numbers = re.findall(r'(?i)(?:Table|Table.})\s+(\d+)', text)
                    if table_numbers:
                        highest_table_number = max(highest_table_number, max(map(int, table_numbers)))
                df.at[i, 'tables'] = highest_table_number
    except Exception as e:
        print(f"Metadata extraction from PDF failed for {df['title'][i]}: {e}")
        non_existent_dates.append(paper_date)
        non_existent_titles.append(df['title'][i])

non_existent = pd.DataFrame({'date': non_existent_dates, 'title': non_existent_titles})
non_existent_write = pd.concat([prev_non_existent, non_existent], ignore_index=True)
non_existent_write.to_csv('datasets/non_existent.csv', index=False)

print('Retrieved missing Metadata')

# For loop to retrieve keywords
if 'keywords' not in df.columns:
    df['keywords'] = None  

for i in tqdm(range(len(df)), desc="Retrieving keywords"):
    try:
        links = df['pdf_link'][i]
        if not links.startswith('http'): links = 'https://' + links
            
        pdf_response = libreq.urlopen(links)
        pdf_file = pdf_response.read()
        pdf_reader = PyPDF2.PdfReader(io.BytesIO(pdf_file))

        keywords = []
        for page in pdf_reader.pages:
            text = page.extract_text()
            text = re.sub(r'\s+', ' ', text) 

            patterns = [
                r'(?i)(?:keyword[s]?|Uniﬁed Astronomy Thesaurus concepts?|key words?|Key words?|Subject headings)[:.]?\s*(.*?)\s*(?=(?:[.;]|\n|$))'
            ]

            for pattern in patterns:
                matches = re.findall(pattern, text, re.DOTALL)
                for match in matches:

                    split_keywords = re.split(r'[;,\n]', match)
                    keywords.extend([kw.strip() for kw in split_keywords if kw.strip()])

        keywords = list(set(keywords))

        stop_phrases = ['1. Introduction', '1 Introduction']
        for stop_phrase in stop_phrases:
            if any(stop_phrase in keyword for keyword in keywords):
                keywords = ' '.join(keywords).split(stop_phrase)[0]
                break

        df.at[i, 'keywords'] = keywords
    except:
        # print(f"cannot retrieve keywords for paper {df['title'][i]}")
        pass
print('Retrieved Keywords')

# Save the data
new_df = pd.concat([prev_df, df], ignore_index=True)
new_df.to_csv('arxiv_papers.csv', index=False)
```

```python
import requests
import os

TOKEN = os.getenv("ADS_API_KEY")
if not TOKEN:
    raise RuntimeError(
        "ADS_API_KEY environment variable is not set. "
        "Set it in the notebook (e.g. `TOKEN = '<your-token>'`) or export it in your shell: "
        "`export ADS_API_KEY='YOUR_TOKEN'` and restart the kernel."
    )

headers = {"Authorization": f"Bearer {TOKEN}"}

params = {
    "q": 'author:"John F Wu"',
    "fl": "author,aff",
    "rows": 1
}

r = requests.get(
    "https://api.adsabs.harvard.edu/v1/search/query",
    headers=headers,
    params=params,
    timeout=10
)

try:
    r.raise_for_status()
except requests.HTTPError:
    print(f"Request failed with status {r.status_code}: {r.reason}")
    print("Response body:", r.text)
    raise

print(r.json())
```

```python
# Print author affiliations
affiliations = r.json().get('response', {}).get('docs', [])
for entry in affiliations:
    print(f"Author: {entry.get('author')}, Affiliations: {entry.get('aff')}")
```

```python
headers = {"Authorization": f"Bearer {TOKEN}"}

title_query = str(paper).replace('"', '\\"')
params = {
    "q": f'title:"{title_query}"',
    "fl": "title,author,aff",
    "rows": 1
}

r = requests.get(
    "https://api.adsabs.harvard.edu/v1/search/query",
    headers=headers,
    params=params
)

data = r.json()['response']['docs']

for paper in data:
    print("\nTITLE:", paper.get("title", ["N/A"])[0])
    for auth, aff in zip(paper.get("author", []), paper.get("aff", [])):
        print(f"  {auth:30} | {aff}")
```

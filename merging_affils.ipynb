```python
def affiliations_need_fix(val):
    """
    Returns True if affiliations is a list and contains None
    """
    if pd.isna(val):
        return True

    if isinstance(val, str):
        try:
            val = ast.literal_eval(val)
        except Exception:
            return True

    if isinstance(val, list):
        return any(v is None or str(v).strip().lower() == None for v in val)

    return True    

to_process = df[df["affiliations"].apply(affiliations_need_fix)].index.tolist()
print(f"Papers needing affiliation fix: {len(to_process)}")
```

```python
# Print pdf_link of all papers needing affiliation fix
for idx in to_process:
    print(df.loc[idx, "pdf_link"])

# Save to text file
with open("papers_needing_affiliation_fix.txt", "w") as f:
    for idx in to_process:
        f.write(f"{df.loc[idx, 'pdf_link']}\n")
```

```python
def create_affiliation_list_from_csv(csv_filepath):
    """
    Reads a CSV file of parsed paper data, filters out incomplete papers,
    and returns a list of semicolon-separated affiliation strings for each valid paper.

    Args:
        csv_filepath (str): The path to the input CSV file.

    Returns:
        list: A list of strings, where each string contains all affiliations
              for a single valid paper, separated by semicolons. Returns None if file is not found.
    """
    try:
        # Load the CSV file. It has no header.
        df = pd.read_csv(csv_filepath, header=None, names=['title', 'author', 'affiliation'])
    except FileNotFoundError:
        print(f"Error: The file '{csv_filepath}' was not found.")
        return None
    except pd.errors.EmptyDataError:
        print(f"Warning: The file '{csv_filepath}' is empty.")
        return []
    except Exception as e:
        print(f"An error occurred while reading the CSV: {e}")
        return None

    # Drop empty rows that were used as separators in the source file
    df.dropna(how='all', inplace=True)
    
    # Normalize missing affiliations: pandas might read empty quoted fields "" as empty strings
    # and completely empty fields ,, as NaN. Let's treat them the same.
    df['affiliation'].fillna('', inplace=True)

    # Identify titles of any paper that has at least one author with a missing affiliation.
    # A missing affiliation is now consistently represented by an empty string ''.
    invalid_titles = df[df['affiliation'] == '']['title'].unique()

    # Filter the DataFrame to keep only the papers where NO author had a missing affiliation
    valid_papers_df = df[~df['title'].isin(invalid_titles)]

    # If no valid papers are left after filtering, return an empty list
    if valid_papers_df.empty:
        return []

    # Group by title, aggregate the affiliations for each paper into a single string, 
    # and then convert the result into a list.
    # The 'affiliation' column for each group (paper) is joined into one string with "; ".
    affiliation_list = valid_papers_df.groupby('title')['affiliation'].apply(lambda x: '; '.join(x)).tolist()

    return affiliation_list
```

```python
def count_authors_with_affiliations(csv_filepath):
    """
    Reads a CSV of parsed paper data, filters out papers with any missing 
    affiliations, and counts the total number of authors from the remaining valid papers.

    Args:
        csv_filepath (str): The path to the input CSV file.

    Returns:
        int: The total count of authors from papers where all authors had an affiliation.
             Returns None if the file cannot be read.
    """
    try:
        # Load the CSV file, which has no header.
        df = pd.read_csv(csv_filepath, header=None, names=['title', 'author', 'affiliation'])
    except FileNotFoundError:
        print(f"Error: The file '{csv_filepath}' was not found.")
        return None
    except pd.errors.EmptyDataError:
        print(f"Warning: The file '{csv_filepath}' is empty.")
        return 0
    except Exception as e:
        print(f"An error occurred while reading the CSV: {e}")
        return None

    # Drop any blank rows used as separators between papers
    df.dropna(how='all', inplace=True)
    
    # Standardize missing affiliation data to empty strings for consistent filtering
    df['affiliation'].fillna('', inplace=True)

    # --- Core Filtering Logic ---
    # 1. Find all unique titles of papers that contain at least one author with a missing affiliation.
    titles_with_missing_affiliations = df[df['affiliation'] == '']['title'].unique()

    # 2. Filter the original DataFrame to keep only papers whose titles are NOT in the list from step 1.
    # This ensures that we only consider papers where every listed author has an affiliation.
    valid_papers_df = df[~df['title'].isin(titles_with_missing_affiliations)]
    
    # 3. The final count is simply the number of rows in this filtered DataFrame.
    author_count = len(valid_papers_df)

    return author_count
```

```python
# --- Configuration ---
# Specify the path to your CSV file here.
csv_file = 'output_2.csv'

# --- Execution ---
total_authors_with_affiliations = count_authors_with_affiliations(csv_file)

# --- Display Result ---
# The function returns None if there was an error reading the file.
if total_authors_with_affiliations is not None:
    print(f"Total number of authors with affiliations (from complete papers): {total_authors_with_affiliations}")
```

```python
def count_all_authors_with_affiliations(csv_filepath):
    """
    Reads a CSV of parsed paper data and counts every author that has a 
    non-empty affiliation string, regardless of whether their co-authors do.

    Args:
        csv_filepath (str): The path to the input CSV file.

    Returns:
        int: The total count of individual authors with a non-empty affiliation.
             Returns None if the file cannot be read.
    """
    try:
        # Load the CSV file, which has no header.
        df = pd.read_csv(csv_filepath, header=None, names=['title', 'author', 'affiliation'])
    except FileNotFoundError:
        print(f"Error: The file '{csv_filepath}' was not found.")
        return None
    except pd.errors.EmptyDataError:
        print(f"Warning: The file '{csv_filepath}' is empty.")
        return 0
    except Exception as e:
        print(f"An error occurred while reading the CSV: {e}")
        return None

    # Drop any blank rows used as separators between papers
    df.dropna(how='all', inplace=True)
    
    # Standardize missing affiliation data to empty strings for consistent filtering
    df['affiliation'].fillna('', inplace=True)

    # --- Core Counting Logic ---
    # The condition `df['affiliation'] != ''` creates a boolean Series (True/False).
    # Summing this Series treats True as 1 and False as 0, effectively counting
    # all rows where the affiliation is not an empty string.
    author_count = (df['affiliation'] != '').sum()
    
    return author_count
```

```python
# --- Configuration ---
# Specify the path to your CSV file here.
csv_file = 'output_2.csv'

# --- Execution ---
total_author_count = count_all_authors_with_affiliations(csv_file)

# --- Display Result ---
if total_author_count is not None:
    print(f"Total number of authors with a non-empty affiliation: {total_author_count}")
```

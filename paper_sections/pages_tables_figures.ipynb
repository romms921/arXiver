```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.dpi': 250})
plt.rcParams['font.sans-serif'] = 'Times New Roman'
import seaborn as sns
import plotly.express as px
import ast
from collections import Counter
```

```python
data = pd.read_csv('/Users/ainsleylewis/Documents/Astronomy/arXiver/2025_Data.csv')
```

```python
# Month Column
data['date'] = pd.to_datetime(data['date'])
data['month'] = data['date'].dt.month
```

```python
# Average number of pages
avg_pages = data['pages'].mean()
print(f"Average number of pages: {avg_pages:.2f}")

# Average number of figures
avg_figures = data['figures'].mean()
print(f"Average number of figures: {avg_figures:.2f}")

# Average number of tables
avg_tables = data['tables'].mean()
print(f"Average number of tables: {avg_tables:.2f}")
```

```python
avg_pages = data['pages'].mean()
med_pages = data['pages'].median()

bins = np.arange(0, 100, 5)
plt.figure(figsize=(12, 8))
plt.hist(data['pages'], bins=bins, edgecolor='white', alpha=0.9, color='darkblue')
plt.axvline(avg_pages, color='red', linestyle='dashed', linewidth=3, label=f'Average Pages: {avg_pages:.2f}')
plt.axvline(med_pages, color='orange', linestyle='dashed', linewidth=3, label=f'Median Pages: {med_pages:.2f}')
plt.title('Distribution of the Number of Pages', fontsize=16)
plt.xlabel('Number of Pages', fontsize=16)
plt.ylabel('Number of Papers', fontsize=16)
plt.legend(fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
avg_tables = data['tables'].mean()
med_tables = data['tables'].median()

bins = np.arange(0, 20, 2)
plt.figure(figsize=(12, 8))
plt.hist(data['tables'], bins=bins, edgecolor='white', alpha=0.9, color='darkblue')
plt.axvline(avg_tables, color='red', linestyle='dashed', linewidth=3, label=f'Average Tables: {avg_tables:.2f}')
plt.axvline(med_tables, color='orange', linestyle='dashed', linewidth=3, label=f'Median Tables: {med_tables:.2f}')
plt.title('Distribution of the Number of Tables', fontsize=16)
plt.xlabel('Number of Tables', fontsize=16)
plt.xticks(bins, fontsize=16)
plt.ylabel('Number of Papers', fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)
plt.show()
```

```python
avg_figures = data['figures'].mean()
med_figures = data['figures'].median()

bins = np.arange(0, 50, 5)
plt.figure(figsize=(12, 8))
plt.hist(data['figures'], bins=bins, edgecolor='white', alpha=0.9, color='darkblue')
plt.axvline(avg_figures, color='red', linestyle='dashed', linewidth=3, label=f'Average Figures: {avg_figures:.2f}')
plt.axvline(med_figures, color='orange', linestyle='dashed', linewidth=3, label=f'Median Figures: {med_figures:.2f}')
plt.title('Distribution of the Number of Figures', fontsize=16)
plt.xlabel('Number of Figures', fontsize=16)
plt.xticks(bins, fontsize=16)
plt.ylabel('Number of Papers', fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: Pages
avg_pages = data['pages'].mean()
med_pages = data['pages'].median()
bins_pages = np.arange(0, 100, 5)

axes[0].hist(data['pages'], bins=bins_pages, edgecolor='black', alpha=0.9)
axes[0].axvline(avg_pages, color='red', linestyle='dashed', linewidth=2, label=f'Average: {avg_pages:.2f}')
axes[0].axvline(med_pages, color='orange', linestyle='dashed', linewidth=2, label=f'Median: {med_pages:.2f}')
axes[0].set_title('Distribution of Paper Lengths')
axes[0].set_xlabel('Number of Pages')
axes[0].set_ylabel('Frequency')
axes[0].legend()

# Plot 2: Tables
avg_tables = data['tables'].mean()
med_tables = data['tables'].median()
bins_tables = np.arange(0, 20, 2)

axes[1].hist(data['tables'], bins=bins_tables, edgecolor='black', alpha=0.9)
axes[1].axvline(avg_tables, color='red', linestyle='dashed', linewidth=2, label=f'Average: {avg_tables:.2f}')
axes[1].axvline(med_tables, color='orange', linestyle='dashed', linewidth=2, label=f'Median: {med_tables:.2f}')
axes[1].set_title('Distribution of Paper Tables')
axes[1].set_xlabel('Number of Tables')
axes[1].set_xticks(bins_tables)
axes[1].set_ylabel('Frequency')
axes[1].legend()

# Plot 3: Figures
avg_figures = data['figures'].mean()
med_figures = data['figures'].median()
bins_figures = np.arange(0, 50, 5)

axes[2].hist(data['figures'], bins=bins_figures, edgecolor='black', alpha=0.9)
axes[2].axvline(avg_figures, color='red', linestyle='dashed', linewidth=2, label=f'Average: {avg_figures:.2f}')
axes[2].axvline(med_figures, color='orange', linestyle='dashed', linewidth=2, label=f'Median: {med_figures:.2f}')
axes[2].set_title('Distribution of Paper Figures')
axes[2].set_xlabel('Number of Figures')
axes[2].set_xticks(bins_figures)
axes[2].set_ylabel('Frequency')
axes[2].legend()

plt.tight_layout()
plt.show()
```
